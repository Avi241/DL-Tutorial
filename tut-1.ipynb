```python
# -*- coding: utf-8 -*-
"""DL-Tutorial.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/13PvMsatZHbGbzkQvHl0U4ayFUVUgLj4Q

# Tutorial - 1
"""

# Basics and loading dataset
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# print(tf.__version__)
mnist = tf.keras.datasets.mnist
(x_train, y_train),(x_test,y_test) = mnist.load_data()
# print(x_train[0])
plt.imshow(x_train[0],cmap=plt.cm.binary) # plt.cm.binary = set colour to B and W
plt.show()
print("Value = ",y_train[0])

x_train = tf.keras.utils.normalize(x_train,axis=1)
x_test = tf.keras.utils.normalize(x_test,axis=1)

plt.imshow(x_train[0],cmap=plt.cm.binary) # plt.cm.binary = set colour to B and W
plt.show()

# Building a model

model = tf.keras.models.Sequential() # Sequential = feedforward no backward dataflow
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation = tf.nn.relu)) # dense = fully connected
model.add(tf.keras.layers.Dense(128, activation = tf.nn.relu)) # Layer-2
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax)) # Output layer

model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
model.fit(x_train,y_train,epochs=10)

# Testing a model
val_loss, val_acc = model.evaluate(x_test,y_test)
print(val_loss,val_acc)

model.save('epic_num_reader.model')
new_model = tf.keras.models.load_model('epic_num_reader.model')
predictions = new_model.predict(x_test)
print(predictions)
print(np.argmax(predictions[100]))
plt.imshow(x_test[100],cmap=plt.cm.binary)
plt.show()


```
